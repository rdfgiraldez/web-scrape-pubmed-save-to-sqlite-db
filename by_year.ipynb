```python
import pandas as pd
import numpy as np
```

```python
articles = pd.read_csv('articles.csv')
articles.head()
```

```python
df = articles[['year','neuron','microglia','astrocyte','glia','oligodendrocyte','schwann']]
df.groupby('year').sum().to_csv('by_year.csv')
```

```python
pd.read_csv('by_year.csv').head()
```

```python
percents = pd.read_csv('by_year.csv')
percents['total']=percents['neuron']+percents['microglia']+percents['astrocyte']+percents['oligodendrocyte']+percents['schwann']
all_list = ['neuron','microglia','astrocyte','oligodendrocyte','schwann']

for a in all_list:
    percents['Percent '+a.title()]=percents[a]/percents['total']
percents.head()
```

```python
percents.to_csv('percents.csv')
```
